## STEP 4 > 사용할 함수 결정하기

### 1. 곡명 : title

```python
title = bugs.find_all("p", {"class" : "title"})
print(title)
```

```python
for index in range(0, len(title)):           #노래 개수만큼 loop 돌릴거야
    titleindex = title[index].find('a').text #리스트형식이니까 [index] 안에 0~노래 개수 들어가면서 차례로 가져오는데 a태그만 가져올거야. 근데 텍스트만 뽑아줘
    print(index+1, titleindex)               # 순위 : 타이틀뽑은거 보여줘
```

### 2. 가수명 : artist

```python
artist = bugs.find_all("p", {"class" : "artist"})
print(artist)
```

```python
for index in range(0, len(artist)):
    artistindex = artist[index].find('a').text
    print(artistindex)
```

### 3. 앨범명 : album

```python
album = bugs.find_all("a", {"class" : "album"})
print(album)
```

```python
for index in range(1, len(album)):     #0부터 시작하지 않는 이유 : a태그에 '앨범'이라는 것이 하나 더 들어가있어서 
    albumindex = album[index].text
    print(albumindex)
```

## STEP 5 > 함수 종합하여 1~100위까지 목록 만들기

```python
for index in range(0, len(title)):       
    titleindex = title[index].find('a').text
    artistindex = artist[index].find('a').text
    albumindex = album[index+1].text
    print(index+1, ' : ', titleindex, artistindex, albumindex)       
```

```python
with open('bugschart', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'artist', 'album'])
        
    for index in range(0, len(title)):       
        titleindex = title[index].find('a').text
        artistindex = artist[index].find('a').text
        albumindex = album[index+1].text
        print(index+1, ' : ', titleindex, artistindex, albumindex)
        writer.writerow([index+1,titleindex, artistindex, albumindex])
```
