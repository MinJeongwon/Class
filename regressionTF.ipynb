## 1. 필요한 패키지 불러오기

```python
import tensorflow as tf # 머신러닝에서 가장 중요한 모듈 tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np # numpy:행렬 형태로 만들기
import matplotlib.pyplot as plt # 그래프에 나타내기
```

## 2. 보스턴 주택 가격 데이터 가져오기

```python
tf.keras.datasets.boston_housing.load_data(path = 'boston_housing.npz')
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.boston_housing.load_data(path = 'boston_housing.npz')
# train과 test 변수에 저장시키기
```

```python
model = Sequential()
model.add(Dense(1, input_shape = [13]))
model.compile(optimizer = 'rmsprop', loss = 'mse')
model.summary()
```

```python
model.fit(x_train, y_train, epochs = 1000)
# epochs = 1 : dataset을 1번만 돌린 것
# fit : 훈련
```

```python
# evaluate는 test한다는 말
model.evaluate(x_test, y_test)
```

```python
print(model.input_shape)
print(model.output_shape)
```

```python
model.predict(np.array([0, 1, 50, 1, 0, 5, -5, 10, 1000, 100, 10, 100, 1]).reshape(1,13))
```
